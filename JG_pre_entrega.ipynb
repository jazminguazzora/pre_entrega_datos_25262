### 3锔 Calidad de los datos

```python
# funci贸n calidad
def calidad(df, nombre, clave=None):
  print(f"***{nombre}***")
  display(df.isna().sum().to_frame("nulos"))
  filas_d = df.duplicated(keep=False).sum()
  print(f"Filas duplicadas (exactas): ", filas_d)
  if clave and clave in df.columns:
    d_key = df[clave].duplicated(keep=False).sum()
    print(f"Filas duplicadas por clave '{clave}': ", d_key)
    if d_key > 0:
      d_ordenados = df[clave][df[clave].duplicated(keep=False)].value_counts().sort_values(ascending=False)
      print("\nTop valores duplicados m谩s frecuentes:")
      display(d_ordenados.head(10))
    else:
        print(f"\nNo se encontraron duplicadas en la clave '{clave}'.")
  else:
    if clave:
      print(f"La clave '{clave}' no existe en el DataFrame.")
    else:
      print("No se indic贸 una clave para verificar duplicados.")
```

```python
# llamado a la funci贸n calidad para ver nulos y duplicados generales y por clave del dataset clientes
calidad(df_clientes, "CLIENTES (original)", "id_cliente")
```

```python
# llamado a la funci贸n calidad para ver nulos y duplicados generales (sin clave) del dataset marketing
calidad(df_marketing, "MARKETING (original)","id_campanha")# llamado a la funci贸n calidad para ver nulos y duplicados generales (sin clave) del dataset clientes# llamado a la funci贸n calidad para ver nulos y duplicados generales (sin clave) del dataset clientes
```

### 4锔 Limpieza de datos

```python
# 1) crear copias independientes para no modificar los datos originales
df_clientes_c = df_clientes.copy()
df_marketing_c = df_marketing.copy()
df_ventas_c = df_ventas.copy()
```

```python
# 2) eliminar filas completamente duplicadas
df_clientes_c = df_clientes_c.drop_duplicates()
df_marketing_c = df_marketing_c.drop_duplicates()
df_ventas_c = df_ventas_c.drop_duplicates()
```

```python
# 3) funci贸n para limpiar texto en columnas tipo string
def normalizar_texto(df):
  for columna in df.select_dtypes(include="object").columns:
    df[columna] = (df[columna].astype(str).str.strip().str.replace(r"[\u200b\t\r\n]", "", regex=True).str.replace(r"\s+", " ", regex=True).str.title())
  return df
```

```python
# 4) normalizaci贸n de fechas
for df in [df_clientes_c, df_marketing_c, df_ventas_c]:
    for columna in df.columns:
        if "fecha" in columna.lower():
            df[columna] = pd.to_datetime(df[columna], errors="coerce",dayfirst=True)
```

```python
# 5) aplicaci贸n de la normalizaci贸n de texto a los df
df_clientes_c = normalizar_texto(df_clientes_c)
df_marketing_c = normalizar_texto(df_marketing_c)
df_ventas_c = normalizar_texto(df_ventas_c)
```

```python
# 6) normalizacion de valores num茅ricos

# campo precio
if "precio" in df_ventas_c.columns:
  df_ventas_c["precio"] = (df_ventas_c["precio"].astype(str).str.replace("$", "",regex=False).str.replace(",", "",regex=False).str.strip())
  df_ventas_c["precio"] = pd.to_numeric(df_ventas_c["precio"], errors="coerce")

# campo cantidad
if "cantidad" in df_ventas_c.columns:
  df_ventas_c["cantidad"] = pd.to_numeric(df_ventas_c["cantidad"], errors="coerce").astype("Int64")
```

```python
# 7) guardad los DataFrame limpios como csv
df_clientes_c.info()
df_clientes_c.to_csv("/content/drive/MyDrive/pre_entrega/dataset/df_clientes_c.csv", index=False)

df_marketing_c.info()
df_marketing_c.to_csv("/content/drive/MyDrive/pre_entrega/dataset/df_marketing_c.csv", index=False)

df_ventas_c.info()
df_ventas_c.to_csv("/content/drive/MyDrive/pre_entrega/dataset/df_ventas_c.csv", index=False)

print(" Archivos guardados: df_clientes_c.csv, df_marketing_c.csv, df_ventas_c.csv" )
```

### 5锔 Transformaci贸n de datos

```python
# detectar productos de alto rendimiento

# funci贸n encontrar_columna
def encontrar_columna(df, candidatos):
  for c in df.columns:
    nombre = c.lower()
    if any(p in nombre for p in candidatos):
      return c
  return None
```

```python
# 1) detectar la columna de producto
prod_col = encontrar_columna(df_ventas_c, ["producto", "id_producto","sku","art铆culo","articulo"])
if prod_col is None:
  raise ValueError("No se encontr贸 ninguna columna de producto en el DataFrame. Renombr谩 una columna a 'producto'")
```

```python
# 2) calcular ingreso por regisro = precio * cantidad (subtotal)
ventas_perf = (df_ventas_c.assign(ingreso = df_ventas_c["precio"] * df_ventas_c["cantidad"]))
```

```python
# 3) agregar m茅tricas por producto
resumen_prod = (ventas_perf.groupby(by=prod_col, dropna=False, as_index=False).agg(ingreso_total=('ingreso', 'sum'), unidades=('cantidad', 'sum'), precio_promedio=('precio', 'mean'), registros=('ingreso', 'size')))
```

```python
# 4) calcular percentil 80 de ingreso_total
p80_ingreso = resumen_prod["ingreso_total"].quantile(q=0.8, interpolation="linear")
```

```python
# 5) filtrar los productos de alto rendimiento y ordenarlos
ventas_top = (resumen_prod.query("ingreso_total >= @p80_ingreso", engine="python").sort_values(by=["ingreso_total","unidades"], ascending=[False, False],na_position="last",ignore_index=True))
```

```python
# 6) mostrar resultados
print(f"Columna de producto detectada: {prod_col}")
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
print(" Productos de ALTO RENDIMIENTO (top 20% por ingreso):")
display(ventas_top.head(20))
```

### 6锔 Agregaci贸n

```python
# 1) detectar columna "categor铆a"
cat_col = encontrar_columna(df_ventas_c, ["categor铆a", "categoria","categoria_producto","rubro"])
if cat_col is None:
  raise ValueError("No se encontr贸 ninguna columna de 'categor铆a' en el DataFrame. Renombr谩 una columna a 'categor铆a'")

# 2) crear columna 'ingreso' en caso de que no exista
if "ingreso" not in df_ventas_c.columns:
  ventas_cat = df_ventas_c.assign(ingreso = df_ventas_c["precio"] * df_ventas_c["cantidad"])
else:
  ventas_cat = df_ventas_c.copy()

# 3) agregaci贸n por categor铆a con groupby + agg
resumen_cat = (ventas_cat.groupby(by=cat_col, dropna=False, as_index=False).agg(ingreso_total=('ingreso', 'sum'), unidades=('cantidad', 'sum'), ventas=('precio', 'size'), precio_promedio=('precio', 'mean')).sort_values(by=["ingreso_total","unidades"], ascending=[False, False],na_position="last",ignore_index=True))

# 4) ticket promedio por venta = ingreso_total / ventas
resumen_cat = resumen_cat.assign(ticket_promedio_por_venta = resumen_cat["ingreso_total"] / resumen_cat["ventas"])

# 5) mostrar resultados
print("Columna de categor铆a detectada:", cat_col)
print(" Resumen por categor铆a:")
display(resumen_cat.head(20))
```
